```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

**Here --> we have skipped the EDA part.**

**step 1. --> Check the Standard Normal Form.**  --> mean = 0 && s.d = 1

```python
## Discriptive statistics of the given data points.
data.describe()
```

```python
## input features only.
data = data.drop(columns=['Outcome'], axis=1)
```

```python
## convert the data into StandardScaler form.which means --> mean = 0 && s.d = 1

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataScaled = scaler.fit_transform(data)
```

```python
dataScaled = pd.DataFrame(dataScaled)
dataScaled.head()
```

```python
dataScaled.shape
```

**Step 2. finding Covariance matrix.**

```python
covarianceMatrix = dataScaled.T @ dataScaled/767
covarianceMatrix
```

**Step. 3 --> Evaluate the eigenValues and eigenVectors using the covariance matrix.**

```python
eigenValues, eigenVectors = np.linalg.eig(covarianceMatrix)
```

```python
## eigenValues = no. of features.

eigenValues    ## take the highest eigenValues for PCA1.
```

```python
eigenVectors
```

```python
## PCA1 data

PCA1_data = dataScaled @ eigenVectors[:, 0]
PCA1_data
```

```python
## PCA2 data

PCA2_data = dataScaled @ eigenVectors[:, 1]
PCA2_data
```

```python
## PCA3 data

PCA3_data = dataScaled @ eigenVectors[:, 7]
PCA3_data
```

**PCA Implementation via sklearn library.**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit_transform(dataScaled)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principleComponent = pca.fit_transform(dataScaled)
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Dimensions')
plt.ylabel('Explained Variance Ratio')
plt.savefig('ScreenPlot.png')
plt.show()
```
